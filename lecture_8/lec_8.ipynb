```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)
```

```python

y_sin = my_sin(x)
y_cos = my_cos(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y_sin, color='b')
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

axes[1].plot(x, y_cos, color='r')
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')

plt.tight_layout()

plt.savefig('sin_cos_plot.pdf')

plt.show()
```
