```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def f(x):
    return np.exp(x)
```

```python
# Compute y = exp(x)
y = f(x)

plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

# Display the plot
plt.show()
```
